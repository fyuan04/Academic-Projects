# Question 1: Central Limit Theorem Simulation (10 pts)
Create a single dimensional pyspark RDD named bernoulli_rdd that contains 10,000 Bernoulli probability distribution data points consisting of integer 0 or 1.  P(0) = P(1) = 0.5.  Use only pyspark RDDs to complete this question.  Create a histogram of the RDD to show that it follows the Bernoulli distribution.

```
# your code here
import matplotlib.pyplot as plt
#from pyspark.sql.functions import rand,round

from pyspark.mllib.random import RandomRDDs
temp_rdd = RandomRDDs.uniformRDD(sc, 10000)

def b_dist_sim(x):
  int_num = int(x*(1000))
  if (int_num % 2) == 0:
    data = 0
  else:
    data = 1
  return data

bernoulli_rdd = temp_rdd.map(b_dist_sim)
plt.hist(bernoulli_rdd.collect(),5)
plt.title("10,000 Bernoulli probability distribution data points")
plt.xlabel("results")
plt.ylabel("Frequency")
plt.show() 
```

# Question 6 (10 pts): 
Join all 4 dataframes together such that the country, year, population, mortality, life expectancy, and fertility columns are collected together in the same dataframe. The join operation should not throw away any data. Name the new dataframe concat_df. Next, delete all rows where the data values (values other than country and year) are all NaN values. Only drop the row if ALL the data values are NaN.

```
# your code here
from pyspark.sql.functions import array, col, explode, lit, struct
from pyspark.sql import DataFrame
from typing import Iterable 

def melt(
        df: DataFrame, 
        ind_col: Iterable[str], value_cols: Iterable[str], 
        var_name: str="Year", value_name: str="value") -> DataFrame:
  
    _to_tidy = array(*(
        struct(lit(c).alias(var_name), col(c).alias(value_name)) 
        for c in value_cols))

    _tmp = df.withColumn("_to_tidy", explode(_to_tidy))

    cols = ind_col + [
            col("_to_tidy")[x].alias(x) for x in [var_name, value_name]]
    return _tmp.select(*cols)
  
  
val_col_names1 = pop_df_1.schema.names[1:]
tidy_pop_df_1 = melt(pop_df_1, ind_col=['Country'], value_cols=val_col_names1)
tidy_pop_df_1 = tidy_pop_df_1.withColumnRenamed("value", "Population")

val_col_names2 = mort_df_1.schema.names[1:]
tidy_mort_df_1 = melt(mort_df_1, ind_col=['Country'], value_cols=val_col_names2)
tidy_mort_df_1 = tidy_mort_df_1.withColumnRenamed("value", "Morttality")

val_col_names3 = life_exp_df_1.schema.names[1:]
tidy_life_exp_df_1 = melt(life_exp_df_1, ind_col=['Country'], value_cols=val_col_names3)
tidy_life_exp_df_1 = tidy_life_exp_df_1.withColumnRenamed("value", "Life_Expectancy")

val_col_names4 = fert_df_1.schema.names[1:]
tidy_fert_df_1 = melt(fert_df_1, ind_col=['Country'], value_cols=val_col_names4)
tidy_fert_df_1 = tidy_fert_df_1.withColumnRenamed("value", "Fertility")

tempdf1 = tidy_pop_df_1.join(tidy_mort_df_1, on = ['Country','Year'], how='outer')
tempdf2 = tempdf1.join(tidy_life_exp_df_1, on = ['Country','Year'], how='outer')
concat_df = tempdf2.join(tidy_fert_df_1, on = ['Country','Year'], how='outer')

concat_df = concat_df.dropna(how='all', subset=['Population', 'Morttality', 'Life_Expectancy','Fertility'])
```
